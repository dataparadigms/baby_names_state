```python
import pandas as pd
from time import *
from ggplot import *
```

```python
df = pd.read_csv('SC.txt', names = ['State', 'Gender', 'Year','Name', 'Freq'])
```

Check for any duplicates, if not 1 then there's a problem to deal with.

```python
df.duplicated(['State','Gender','Year','Name']).nunique()
```

```python
df['Rank'] = df.groupby(['State','Year','Gender'])['Freq'].rank(ascending=False, method='min')
```

```python
df['RFreq'] = df.groupby(['Year', 'Gender'])['Freq'].apply(lambda x: x.astype(float)/x.sum())
```

```python
df = df.sort(['Gender', 'Year', 'Freq'], ascending = [1,1,0])
df.head()
```

```python
df.tail()
```

```python
p = ggplot(df[df['Rank'] == 1], aes('Year','RFreq', colour='Gender')) + \
        geom_point() + geom_line() + \
        ggtitle('Relative Frequency of Top Ranked Name in SC') + \
        labs(x = 'Year', y = 'Relative Frequency') + \
        theme_bw()
print(p)
```

```python
for gend in df['Gender'].unique():
    p = ggplot(df[(df['Rank'] == 1) & (df['Gender'] == gend)], aes('Year','RFreq', colour = 'Name')) +\
            geom_point() + \
            ggtitle('Gender Specific (' + gend + ') Top Relative Frequency in SC') + \
            labs(x = 'Year', y = 'Relative Frequency') + \
            theme_bw()
    print(p)
```
